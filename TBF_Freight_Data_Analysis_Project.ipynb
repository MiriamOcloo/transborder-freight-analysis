```python
import glob

monthly_zip_files = glob.glob('/content/BTS_data/data/**/*.zip', recursive=True)

print(f"Found {len(monthly_zip_files)} ZIP files:")
for file in monthly_zip_files:
    print(file)
```

```python
import zipfile
import os

extracted_data_path = '/content/all_csvs'
os.makedirs(extracted_data_path, exist_ok=True)

for zip_path in monthly_zip_files:
    try:
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(extracted_data_path)
        print(f" Extracted: {zip_path}")
    except Exception as e:
        print(f" Failed to extract {zip_path}: {e}")
```

```python
import glob

csv_files = glob.glob(os.path.join(extracted_data_path, '**/*.csv'), recursive=True)

print(f"Found {len(csv_files)} CSV files:")
for file in csv_files:
    print(file)
```

```python
import pandas as pd

df_list = []

for file in csv_files:
    try:
        df = pd.read_csv(file)
        if not df.empty:
            df_list.append(df)
        else:
            print(f" Skipped empty file: {file}")
    except Exception as e:
        print(f" Error reading {file}: {e}")

if df_list:
    combined_df = pd.concat(df_list, ignore_index=True, sort=False)
    print(" Combined dataframe shape:", combined_df.shape)
    combined_df.head()
else:
    print("No valid CSV files found.")
```

**1. Top Modes of Transportation by Total Value and Shipping Weight**

```python
top_modes = combined_df.groupby('DISAGMOT')[['VALUE', 'SHIPWT']].sum().sort_values(by='VALUE', ascending=False)
top_modes
```

```python
top_modes.plot(kind='bar', figsize=(10,6), title='Top Transportation Modes by Value and Shipping Weight')
```

```python
disagmot_map = {
    1: 'Vessel',
    3: 'Air',
    4: 'Mail (USPS)',
    5: 'Truck',
    6: 'Rail'
}
```

```python
combined_df['DISAGMOT_DESC'] = combined_df['DISAGMOT'].map(disagmot_map)
```

```python
mode_summary = combined_df.groupby('DISAGMOT')[['VALUE', 'SHIPWT']].sum()

mode_summary.index = mode_summary.index.map(disagmot_map).fillna('Other / Unknown')

mode_summary = mode_summary.sort_values(by='VALUE', ascending=False)

mode_summary.plot(kind='bar', figsize=(10,6), title='Top Transportation Modes by Trade Value and Shipping Weight')
```

**3. How does the trade value and shipping weight vary across different months and years?**

```python
import matplotlib.pyplot as plt
import seaborn as sns


monthly_trends = combined_df.groupby(['YEAR', 'MONTH'])[['VALUE', 'SHIPWT']].sum().reset_index()

monthly_trends['DATE'] = pd.to_datetime(monthly_trends['YEAR'].astype(int).astype(str) + '-' + monthly_trends['MONTH'].astype(int).astype(str) + '-01')

plt.figure(figsize=(14, 5))
sns.lineplot(data=monthly_trends, x='DATE', y='VALUE', marker='o', label='Trade Value')
plt.title('Monthly Trade Value Over Time')
plt.xlabel('Date')
plt.ylabel('Total Trade Value (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 5))
sns.lineplot(data=monthly_trends, x='DATE', y='SHIPWT', marker='o', color='orange', label='Shipping Weight')
plt.title('Monthly Shipping Weight Over Time')
plt.xlabel('Date')
plt.ylabel('Total Shipping Weight')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

**4. Whats the difference in total value, shipping weight, and freight charges between domestic and foreign produced merchandise?**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_q4 = combined_df.groupby('DF')[['VALUE', 'SHIPWT', 'FREIGHT_CHARGES']].sum().reset_index()

df_q4['DF'] = df_q4['DF'].map({1: 'Domestic', 2: 'Foreign'})

df_q4_melted = df_q4.melt(id_vars='DF', var_name='Metric', value_name='Total')

plt.figure(figsize=(12, 6))
sns.barplot(data=df_q4_melted, x='Metric', y='Total', hue='DF')
plt.title('Comparison of Value, Shipping Weight, and Freight Charges\nBetween Domestic and Foreign Merchandise')
plt.ylabel('Total Value / Weight / Charges')
plt.xlabel('Metric')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**5. Which mode of transportation incurs the highest average freight charges?**

```python
mode_map = {
    1: 'Vessel',
    3: 'Air',
    4: 'Mail',
    5: 'Truck',
    6: 'Rail'
}

df_q5 = combined_df.groupby('DISAGMOT')['FREIGHT_CHARGES'].mean().reset_index()

df_q5['Transport_Mode'] = df_q5['DISAGMOT'].map(mode_map)

df_q5 = df_q5.dropna(subset=['Transport_Mode'])

df_q5 = df_q5.sort_values(by='FREIGHT_CHARGES', ascending=False)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(data=df_q5, x='FREIGHT_CHARGES', y='Transport_Mode', hue='Transport_Mode', palette='viridis', legend=False)
plt.title('Average Freight Charges by Mode of Transportation')
plt.xlabel('Average Freight Charges')
plt.ylabel('Mode of Transportation')
plt.tight_layout()
plt.show()
```

**6. Are there notable trends in freight movement across different U.S. border states over time?**

```python
df_q6 = combined_df.groupby(['YEAR', 'USASTATE'])[['SHIPWT', 'VALUE']].sum().reset_index()

df_q6 = df_q6[df_q6['YEAR'] >= 2020]

plt.figure(figsize=(14, 6))
sns.lineplot(data=df_q6, x='YEAR', y='SHIPWT', hue='USASTATE', palette='tab10', marker='o')
plt.title('Shipping Weight Trends by U.S. Border State (2020+)')
plt.xlabel('Year')
plt.ylabel('Shipping Weight')
plt.legend(title='State', bbox_to_anchor=(1.15, 1), loc='upper left')
plt.grid(True)
plt.xticks(sorted(df_q6['YEAR'].dropna().astype(int).unique()))
plt.tight_layout()
plt.subplots_adjust(top=0.9, bottom=0.2)
plt.show()

plt.figure(figsize=(14, 6))
sns.lineplot(data=df_q6, x='YEAR', y='VALUE', hue='USASTATE', palette='tab10', marker='o')
plt.title('Trade Value Trends by U.S. Border State (2020+)')
plt.xlabel('Year')
plt.ylabel('Total Value')
plt.legend(title='State', bbox_to_anchor=(1.15, 1), loc='upper left')
plt.grid(True)
plt.xticks(sorted(df_q6['YEAR'].dropna().astype(int).unique()))
plt.tight_layout()
plt.subplots_adjust(top=0.9, bottom=0.2)
plt.show()
```

**7. What are the most common trade types (import/export), and how do their values and weights compare over time?**

```python
trade_type_map = {1: 'Export', 2: 'Import'}
combined_df['TRADE_TYPE'] = combined_df['TRDTYPE'].map(trade_type_map)

df_q7 = combined_df.groupby(['YEAR', 'TRADE_TYPE'])[['VALUE', 'SHIPWT']].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=df_q7, x='YEAR', y='VALUE', hue='TRADE_TYPE', palette='Set2')
plt.title('Trade Value Over Time by Trade Type')
plt.xlabel('Year')
plt.ylabel('Total Trade Value')
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(data=df_q7, x='YEAR', y='SHIPWT', hue='TRADE_TYPE', palette='Set2')
plt.title('Shipping Weight Over Time by Trade Type')
plt.xlabel('Year')
plt.ylabel('Total Shipping Weight')
plt.grid(True)
plt.tight_layout()
plt.show()
```
